### 计算半年线：

```python
bannian=[]
sum=0
for i in total:
    sum+=((total[i].iloc[120,0]-total[i].iloc[0,0])/(total[i].iloc[0,0]*120))
bannianxian=sum/len(total)
```

```python
def bannian(day,period):

    shoupan=[]
    diyu=[]
    for i in total:
    #     print (type(total[i]))
        try:
            a=(total[i].iloc[day-1,0]-total[i].iloc[day-10,0])/(total[i].iloc[day-10,0]*period)
        except:
            pass
    #     print (a)
        shoupan.append(a)
    shoupan.sort(reverse=True)
    point=shoupan[29]
    gupiao={}
    for i in total:
        try:
            if (total[i].iloc[day-1,0]-total[i].iloc[day-10,0])/(total[i].iloc[day-10,0]*period)>=point:
                gupiao[i]=(total[i].iloc[day-1,0]-total[i].iloc[day-10,0])/(total[i].iloc[day-10,0]*period)
        except:
            pass
    qingkuang=pd.DataFrame(columns=['持有平均收益率','是否高于半年线'],index=range(len(gupiao)))
    for v,i in enumerate(gupiao):
        a=(total[i].iloc[day+5,0]-total[i].iloc[day,0])/(total[i].iloc[day,0]*period)
        qingkuang.iloc[v,0]=a
        if (a)<bannianxian:
            qingkuang.iloc[v,1]=False
        else:
            qingkuang.iloc[v,1]=True
    gupiao=pd.DataFrame(gupiao,index=range(len(gupiao)))
# 股票后10天收益率高于半年线则为TRUE，低于则为FALSE。
    print ('前30的股票代码持有期的收盘价变化：')
    print (qingkuang)
    return (qingkuang)
```
