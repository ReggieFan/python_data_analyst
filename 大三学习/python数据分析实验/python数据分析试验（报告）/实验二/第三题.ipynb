```python
import jieba
import jieba.posseg as pseg       #词性标注
import jieba.analyse as anls        #关键词提取
from collections import Counter
tangshi_text = open('D://唐诗三百首完整版.txt',encoding='UTF-8').read()
print (len(tangshi_text))
#print (tangshi_text)
tangshi_words = [x for x in jieba.cut(tangshi_text) if len(x) >= 2]
c = Counter(tangshi_words).most_common(30)                #Counter有方法去获取最频繁的前30个数据
print (c)
for x, w in anls.extract_tags(tangshi_text, topK=20, withWeight=True):
    print('%s %s' % (x, w))
```

# 统计所有作者诗歌数量，以及总数量

```python
items = tangshi_text.split("\n\n")
total=[]
num=[]
for i,value in enumerate(items):                 # 根据唐诗三百首的文本内容格式，取出诗歌作者名字，然后统计
    name=items[i].split("\n")[1][3:6]
    total.append(name)
length=len(total)
ranks={}
for i in total:
    if total.count(i)>=1:
        ranks[i] = total.count(i)        
for item in sorted(ranks.items(), key=lambda item:item[1],reverse=True):
    num.append(item[0])
ranks ={}
for i,value in enumerate(num):
    ranks[value]=i+1
total=Counter(total)
```

```python
zuozhe=str(input("请输入作者名字，可以得出作者诗歌数量,数量占比，排行："))
print ('作者的诗歌数量：'+str(total[zuozhe])+'\n数量占比：'+str(total[zuozhe]/length)+'\n排行：'+str(ranks[zuozhe]))
```
